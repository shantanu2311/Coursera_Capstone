```python
import requests # library to handle requests
import pandas as pd # library for data analsysis
import numpy as np # library to handle data in a vectorized manner

!pip install geopy 
from geopy.geocoders import Nominatim # module to convert an address into latitude and longitude values

```

```python
import urllib.request
url = "https://indiamapia.com/Gurgaon.html"
page = urllib.request.urlopen(url)

from bs4 import BeautifulSoup 
soup = BeautifulSoup(page, "html5lib")
```

```python
tables=soup.find_all("table") 

for table in tables:
    head = table.find_all('h2')
    headings = [h2.text.strip() for th2 in head]
    if headings[:4] == ['Gurgaon', ' Pin', 'Code', 'List']:
        break
table
```

```python
A=[]
B=[]

for row in table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==2:
        A.append(cells[0].findAll(text=True))
        B.append(cells[1].findAll(text=True))

df=pd.DataFrame(A,columns=['Area'])
df['Pincode']=B
df.head()
```

```python
df['Pincode'] = df['Pincode'].str[0]
df
```

```python
import pgeocode
lis  = df.loc[:,"Pincode"]
lis2 = lis.tolist()
nominatim = pgeocode.Nominatim('IN') #Country code
```

```python
df2 = nominatim.query_postal_code(lis2)
df2.head()
```

```python
df2.rename(columns = {'postal_code':'Pincode'},inplace=True)
df3= pd.merge(df,df2,on ='Pincode', how ='inner')
df3.head()
```

```python
df3=df3.groupby('Pincode')
df3.head()
```

```python
df3=df3.drop(df3[2:7],axis=1)
df3.head()
```

```python
df4 = df3.filter(['Area','Pincode','community_name','latitude','longitude'], axis=1)
df4.head()
```

```python
df5= df4.drop_duplicates(subset='Area', keep='first')
df4
```

```python
df5=df5.dropna()
df5 = df5.reset_index(drop=True)
df5
```

```python
df6=df5.sort_values("Pincode")
df6 = df6.reset_index(drop=True)
df6["community_name"].nunique()
```

```python
df7=df6.filter(['community_name','latitude','longitude'],axis=1)
```
